### 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
from bokeh.models import ColumnDataSource
sizes = np.random.randint(10, 30, size=100)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=100)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

output_notebook()

p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

show(p)
```

### 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource

fruits = ["Apples", "Oranges", "Bananas", "Pears"]
counts = [5, 3, 4, 2]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
output_notebook()

p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color='navy', alpha=0.7)

p.xgrid.grid_line_color = None

show(p)
```

### 4. Create a Bokeh histogram to visualize the distribution of the given data.data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_notebook()
p = figure(title="Histogram", background_fill_color="beige")

# Add a quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="purple", line_color="black", alpha=0.5)

# Show the plot
show(p)
```

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.palettes import Viridis256
output_notebook()
mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None, tools="")

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
