```python
# 1 What is a parameter
A parameter is a variable listed inside the parentheses in the function definition.
It acts as a placeholder to receive values (arguments) when the function is called.
During the function call, the argument values are assigned to the corresponding parameters,
allowing the function to work with those specific values.
```

```python
# 7 What do you mean by training and testing a dataset?
The training dataset is the portion of our data used to train the machine learning model.
The model learns patterns and relationships from this data by adjusting its internal parameters to make accurate predictions.
It's like showing the model many examples with answers so it can learn the underlying rules.


Testing dataset  is a special set of data that the model hasn't seen during training.
We use the trained model to make predictions on the testing dataset.
We compare the model's predictions to the actual answers in the testing dataset to see how well it did.
```

```python
# 15 What is causation? Explain difference between correlation and causation with an example
Causation means one thing directly causes the other to happen.

Correlation means two things happen together, but one doesn’t necessarily cause the other.
Example:
Correlation: Ice cream sales and swimming both increase in summer. But eating ice cream doesn’t make people swim.
Causation: Smoking causes lung cancer. Smoking directly harms the lungs, leading to cancer.
```

```python
# 18 What does model.fit() do?
The model.fit() method in scikit-learn is used to train a machine learning model on the provided data.
This method adjusts the model's parameters (e.g., weights in a linear model) based on the input data so that the model can make accurate predictions.
Training the model: When you call model.fit(X, y), the model learns from the input data X (features) and the target values y (labels or outcomes).
Adjusting parameters: The method adjusts the internal parameters of the model to minimize the difference between the model's predictions and
the actual target values, usually by minimizing a loss function.
```

```python
# 19 What does model.predict() do? What arguments must be given?
Prediction: After training the model with fit(), predict() takes new input data and applies the model’s learned patterns to make predictions.
Output: It returns the predicted target values (such as predicted class labels or continuous values) based on the provided input features.

Arguments that must be given:
X_new (New input data):
This is the new data for which we want to make predictions.
It should be in the same format as the training data used in model.fit(). Typically, X_new is a 2D array (a list of lists, numpy array, or pandas DataFrame), where:
Each row represents one sample (observation).
Each column represents one feature (attribute) of the sample.
```
